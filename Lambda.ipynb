```python
Max=lambda a,b:a if(a>b) else b
print(Max(4,2))
```

```python
list=[[2,3,4],[1,24,16,64],[3,6,9,12]]
sortlist=lambda x:[sorted(i) for i in x]
print(sortlist(list))
s=lambda x:[y[-2] for y in sortlist(x)]
print(s(list))
```

```python
from functools import reduce
li=[5,8,10,20,50,100]
sum=reduce(lambda x,y:x+y,li)
print(sum)
```
