```python
def logical_or(input1, input2):
    """Implements a logical OR gate."""
    return input1 or input2

print("Logical OR Gate Truth Table:")
print("Input 1 | Input 2 | Output")
print("------------------------")

# Test cases for the OR gate
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

for i1, i2 in inputs:
    output = logical_or(i1, i2)
    print(f"{i1:^7} | {i2:^7} | {output:^6}")
```

```python
import matplotlib.pyplot as plt

# Data from the logical OR gate truth table
input_combinations = [(0, 0), (0, 1), (1, 0), (1, 1)]
or_outputs = [0, 1, 1, 1] # Corresponding outputs for the OR gate

# Create labels for the x-axis
x_labels = [f"({i1}, {i2})" for i1, i2 in input_combinations]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(x_labels, or_outputs, color=['red', 'green', 'green', 'green'])

# Add title and labels
plt.title('Logical OR Gate Output')
plt.xlabel('Input Combination (Input 1, Input 2)')
plt.ylabel('Output')
plt.ylim(-0.1, 1.1) # Set y-axis limits to clearly show 0 and 1
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define the Perceptron for OR Gate ---
# The OR gate outputs 0 only when both inputs are 0, otherwise it outputs 1.
# A simple perceptron can model this.
# We can set weights and bias manually, or train it for simplicity.
# For OR gate, weights could be (1, 1) and bias -0.5

class Perceptron:
    def __init__(self, weights, bias, activation_function):
        self.weights = np.array(weights)
        self.bias = bias
        self.activation_function = activation_function

    def predict(self, inputs):
        # Weighted sum of inputs plus bias
        z = np.dot(inputs, self.weights) + self.bias
        # Apply activation function
        return self.activation_function(z)

# Step activation function (binary classifier)
def step_activation(x):
    return 1 if x >= 0 else 0

# Define the OR gate perceptron
or_perceptron = Perceptron(weights=[1, 1], bias=-0.5, activation_function=step_activation)

# Input data for OR gate
or_inputs = np.array([
    [0, 0], # Output should be 0
    [0, 1], # Output should be 1
    [1, 0], # Output should be 1
    [1, 1]  # Output should be 1
])

# Corresponding labels (expected outputs)
or_labels = np.array([0, 1, 1, 1])

# --- 2. Visualize the Perceptron's Decision Boundary ---
plt.figure(figsize=(8, 6))

# Plot data points
for i, input_point in enumerate(or_inputs):
    if or_labels[i] == 0:
        plt.scatter(input_point[0], input_point[1], color='red', marker='o', s=100, label='Class 0' if i == 0 else '')
    else:
        plt.scatter(input_point[0], input_point[1], color='blue', marker='x', s=100, label='Class 1' if i == 1 else '')

# Plot the decision boundary
# The decision boundary is defined by w1*x1 + w2*x2 + bias = 0
# For the OR perceptron: 1*x1 + 1*x2 - 0.5 = 0  => x2 = -x1 + 0.5

x_boundary = np.linspace(-0.5, 1.5, 100)
y_boundary = -or_perceptron.weights[0]/or_perceptron.weights[1] * x_boundary - or_perceptron.bias/or_perceptron.weights[1]

plt.plot(x_boundary, y_boundary, 'k--', label='Decision Boundary')

plt.title('Perceptron for OR Gate with Decision Boundary')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.xlim([-0.2, 1.2])
plt.ylim([-0.2, 1.2])
plt.grid(True)
plt.axvline(0, color='gray', linewidth=0.5)
plt.axhline(0, color='gray', linewidth=0.5)
plt.legend()
plt.show()
```
