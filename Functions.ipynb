### More Function Examples

```python
def calculate_area(length, width=10):
  """Calculates the area of a rectangle. Width defaults to 10 if not provided."""
  return length * width

print(f"Area with length 5 and default width: {calculate_area(5)}")
print(f"Area with length 5 and width 20: {calculate_area(5, 20)}")
```

```python
def sum_all_numbers(*args):
  """Sums an arbitrary number of positional arguments."""
  total = 0
  for num in args:
    total += num
  return total

print(f"Sum of 1, 2, 3: {sum_all_numbers(1, 2, 3)}")
print(f"Sum of 10, 20, 30, 40, 50: {sum_all_numbers(10, 20, 30, 40, 50)}")

#*args collects all non-keyword (positional) arguments into a tuple.
```

```python
def print_user_info(**kwargs):
  """Prints user information provided as keyword arguments."""
  for key, value in kwargs.items():
    print(f"{key.replace('_', ' ').title()}: {value}")

print_user_info(name="Alice", age=30, city="New York")
print_user_info(first_name="Bob", last_name="Smith", occupation="Engineer")

#**kwargs collects all keyword arguments into a dictionary.
```

```python
def multiply(a: int, b: int) -> int:
  """Multiplies two integers, demonstrating type hints."""
  return a * b

print(f"Multiplication of 7 and 8: {multiply(7, 8)}")
# Type hints are suggestions, Python doesn't enforce them at runtime by default.
print(f"Multiplication of 'hello' and 3: {multiply('hello', 3)}")
```

```python
def factorial(n: int) -> int:
  """Calculates the factorial of a non-negative integer recursively."""
  if n == 0 or n == 1:
    return 1
  else:
    return n * factorial(n-1)

print(f"Factorial of 5: {factorial(5)}") # 5 * 4 * 3 * 2 * 1 = 120
print(f"Factorial of 0: {factorial(0)}")
```

```python
square_lambda = lambda x: x * x
add_lambda = lambda x, y: x + y

print(f"Square of 9 using lambda: {square_lambda(9)}")
print(f"Addition of 15 and 25 using lambda: {add_lambda(15, 25)}")
```

```python
def get_min_max(numbers):
  """Returns the minimum and maximum values from a list as a tuple."""
  if not numbers:
    return None, None
  return min(numbers), max(numbers)

nums = [4, 1, 8, 3, 9, 2]
minimum, maximum = get_min_max(nums)
print(f"Original list: {nums}")
print(f"Minimum: {minimum}, Maximum: {maximum}")

empty_list = []
min_val, max_val = get_min_max(empty_list)
print(f"Empty list: {empty_list}")
print(f"Minimum: {min_val}, Maximum: {max_val}")
```

```python
def count_up_to(limit):
  """A generator function that yields numbers from 1 up to a limit."""
  n = 1
  while n <= limit:
    yield n
    n += 1

print("Counting up to 5:")
for num in count_up_to(5):
  print(num)

# Generators can be converted to lists
list_from_generator = list(count_up_to(3))
print(f"List from generator (up to 3): {list_from_generator}")
```
