```python
max_num = None

choice = "yes"
while choice == "yes":
    num = int(input("Enter number: "))

    if max_num is None or num > max_num:
        max_num = num

    choice = input("More? (yes/no): ")

print("Maximum number is:", max_num)
```

```python
class employ:
  def __init__(self,name,age,salary):
    self.name=name
    self.age=age
    self.salary=salary
  def display(self):
    print(f"employ {self.name} {self.age} {self.salary}")
  def up(self):
    self.salary=self.salary+100;

emplyee = [
    employ("ram",20,1000),
    employ("syam",21,2000)
]
for emp in emplyee:
  print(f"Name:{emp.name},Age:{emp.age}")
```
